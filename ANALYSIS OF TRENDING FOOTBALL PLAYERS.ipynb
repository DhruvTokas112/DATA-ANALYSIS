```python
# PLAYER WITH THE HIGHEST WAGE
```

```python
def convert_wage_to_numeric(wage_str):
    value_str = wage_str.replace('€', '').replace('K', '').replace('M', '')
    unit = wage_str[-1]
    if unit == 'M':
        return float(value_str) * 1000000
    elif unit == 'K':
        return float(value_str) * 10000
    else:
        return float(value_str)
data['Wage (Numeric)'] = data['Wage'].apply(convert_wage_to_numeric)
player_with_highest_wage = data.loc[data['Wage (Numeric)'].idxmax()]
print(player_with_highest_wage[['Player_name', 'Wage']])
```

```python
# MOST VALUABLE PLAYER
```

```python
def convert_value_to_numeric(value_str):
    if value_str.endswith('M'):
        return float(value_str.replace('€', '').replace('M', '')) * 1000000
    elif value_str.endswith('K'):
        return float(value_str.replace('€', '').replace('K', '')) * 1000

    else:
        return float(value_str.replace('€', ''))
data['Value (Numeric)'] = data['Value'].apply(convert_value_to_numeric)
most_valuable_player = data.loc[data['Value (Numeric)'].idxmax()]
print(most_valuable_player[['Player_name', 'Value']])
```

```python
#NATIONAL TEAM RANKING 
```

```python
avg_overall = data.groupby('National_team')['Overall'].mean().reset_index()
sorted_teams = avg_overall.sort_values(by='Overall', ascending=False)
print(sorted_teams)
```

```python
# CLUBS RANKING
```

```python
club_avg_ratings = data.groupby('Current_club')['Overall'].mean().sort_values(ascending=False)
print(club_avg_ratings)
```
